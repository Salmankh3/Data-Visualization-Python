-------------------------

Plotting More Than One Data Points

```python
np.random.seed(10)
data1 = np.random.normal(80,10,200)
data2 = np.random.normal(50,20,200)
data3 = np.random.normal(30,30,200)
data4 = np.random.normal(10,40,200)

new_data = [data1,data2,data3,data4]

fig = plt.figure(figsize=(10,7))

ax = fig.add_axes([0,0,1,1])

box = ax.boxplot(new_data)

plt.show()
```

---------------------------------------------------------------------

RUNS , OVERS , RUNRATES SCENERIO

```python
runs = np.array([150,140,133,166,177,108,123,173,145,190,144,177,183,166,119])
overs = np.array([20,19.33,18.67,19.83,20,12.5,16.16,18.33,17.5,19.5,14,18.16,20,19.83,10.5])

r_rates = runs / overs

print("Runs : ",runs)
print("\nOvers : ",overs)
print("\nRun Rates : ",r_rates)
```

```python
# Plotting for Runs
fig = plt.figure(figsize=(10,7))
plt.boxplot(runs)
plt.show()
```

```python
# Plotting For Overs
plt.boxplot(overs)
plt.show()
```

```python
# Plotting For Run Rates
plt.boxplot(r_rates)
plt.show()
```

```python
# Plotting Combination of Data (Note: You can plot either two of them or you can also plot all of them at a time)

r_data = [runs,overs]

fig = plt.figure(figsize=(10,7))

ax = fig.add_axes([0,0,1,1])

b_plot = ax.boxplot(r_data)

plt.show()
```
