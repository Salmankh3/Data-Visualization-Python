# MATPLOTLIB LIBRARY

- Matplotlib is one of the most popular Python packages used for Data Visualization
- Created by John D. Hunter in 2003
- It is open-source
- Almost written in Python except of some segments that are written in C , Object-C and JavaScript
- Latest version is 3.3.4

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Plotting two variables x , y
x = np.array([1,2,3,4])
y = np.array([10,20,30,40])

plt.plot(x,y)
plt.show()
```

```python
# Changing line color to green
plt.plot(x,y,color="g")
plt.show()
```

```python
# Plotting single array values with red line
a = np.array([9,3,13,18,4,8])
plt.plot(a, color="r")
plt.show()
```

```python
# Plotting random temperature in a month day by day
days = list(range(1,31))
temp = [random.randint(20,40) for b in range(30)]

plt.plot(days,temp)
plt.show()
```

```python
# Plotting exponent function
s = np.arange(1,6)
e = np.exp(s)

plt.plot(s,e,color="black")
plt.show()
```
