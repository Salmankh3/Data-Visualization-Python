# BARS

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["A","B","C","D"])
y = np.array([12,15,10,7])

plt.bar(x,y)
plt.show()
```

```python
fruit = np.array(["Apples","Bananas","Oranges"])
price = np.array([60,40,25])

plt.bar(fruit,price)
plt.show()
```

```python
# Horizontal Bar
x = np.array(["A","B","C","D"])
y = np.array([12,15,10,7])

plt.barh(x,y)
plt.show()
```

```python
# Changing Color of Bars 
fruit = np.array(["Apples","Bananas","Oranges"])
price = np.array([60,40,25])

plt.bar(fruit,price,color="orange")
plt.show()
```

```python
# Setting width of Vertical Bars
fruit = np.array(["Apples","Bananas","Oranges"])
price = np.array([60,40,25])

plt.bar(fruit,price,color="g",width=0.4)
plt.show()
```

```python
# Setting height of Horizontal Bars
x = np.array(["A","B","C","D"])
y = np.array([12,15,10,7])

plt.barh(x,y, color="k", height=0.2)
plt.show() 
```

SCENERIO OF GENDER POPULATION IN GROUPS

```python
men = [24,21,27,29,40,23]
women = [27,20,18,15,30,17]
groups = np.arange(6)

plt.bar(groups,men, color="r", width=.35)
plt.bar(groups,women, color="b", width=.35, bottom = men)
plt.xlabel("Groups")
plt.ylabel("Population")
plt.title("Gender Population In Groups")
plt.xticks(groups,("G1","G2","G3","G4","G5","G6"))
plt.yticks(np.arange(0,81,10))
plt.legend(labels=["Men","Women"])
plt.show()
```

SCENERIO OF STUDENTS IN A GROUP HAVING DIFFERENT DOMAINS

```python
std_dom1 = np.array([4,6,10,12,5])
std_dom2 = np.array([7,2,3,1,5])
std_dom3 = np.array([4,7,2,2,5])
groups = np.arange(5)

plt.bar(groups,std_dom1, color="b", width=.35)
plt.bar(groups,std_dom2, color="g", width=.35, bottom = std_dom1)
plt.bar(groups,std_dom3, color="k", width=.35, bottom = std_dom1 + std_dom2)
plt.xlabel("Groups")
plt.ylabel("Students Of Different Domains")
plt.title("STUDENTS IN A GROUP HAVING DIFFERENT DOMAINS")
plt.xticks(groups,("G1","G2","G3","G4","G5"))
plt.yticks(np.arange(0,25,2))
plt.legend(labels=["AI","BC","CC"])
plt.show()
```
