```python
import matplotlib.pyplot as plt
import numpy as np
```

LINES

```python
x = np.array([10,12,11,8,7,9])
plt.plot(x, linestyle="dotted")
plt.show()
```

```python
x = np.array([10,12,11,8,7,9])
plt.plot(x, linestyle="-.", color="green")
plt.show()
```

```python
x = np.array([10,12,11,8,7,9])
y = np.array([12,13,18,15,11,9])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x = np.array([10,12,11,8,7,9])
y = np.array([12,13,18,15,11,9])
plt.plot(x, linestyle="-.", color="green", linewidth=3)
plt.plot(y, linestyle="dotted", color="maroon", linewidth=3)
plt.show()
```

MARKERS

```python
a = np.array(["Salman","Uzaif","Talha","Nabeel"])
plt.plot(a, marker="o")
plt.show()
```

```python
a = np.array(["Salman","Uzaif","Talha","Nabeel"])
b = np.array([80,75,90,95])
plt.plot(a,b, marker="*", ms=14, mec="r", mfc="g")
plt.show()
```

```python
a = np.array(["Vitz","Swift","Alto","Mehran"])
b = np.array([7.5,10.3,6.2,4.8])
plt.plot(a,b, marker="D", color="black",ms=10,mec="green",mfc="white")
plt.show()
```

```python
# marker / line / color

a = np.array([10,6,5,2,7,9,4])
b = np.arange(1,8)
plt.plot(b,a,"o:r")
plt.show()
```

```python
a = np.array([10,6,5,2,7,9,4])
b = np.arange(1,8)
plt.plot(b,a,"D-.g",mec="black",ms=10)
plt.show()
```

```python
# graph of linespace for every 2x interval
n = 256

a = np.linspace(-np.pi, np.pi, n, endpoint=True)
b = np.sin(2*a)

plt.plot(a,b, color="blue", alpha=1.5)
plt.fill_between(a, b, 0, color="blue", alpha=0.3)
plt.show()
```

```python
n = 10

a = np.linspace(-np.pi, np.pi, n, endpoint=True)
b = np.sin(2*a)

plt.plot(a,b, color="red", alpha=1.5)
plt.fill_between(a, b, 1, color="red", alpha=0.2)
plt.show()
```
