# SCATTER PLOTS

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# SIMPLE SCATTER

x = np.array([23,12,15,17,13,11,21,18,20,22,26])
y = np.array([44,45,49,55,46,49,54,65,55,53,46])

plt.scatter(x,y)
plt.show()
```

```python
# MULTIPLE SCATTER

x = np.array([23,12,15,17,13,11,21,18,20,22,26])
y = np.array([44,45,49,55,46,49,54,65,55,53,46])

plt.scatter(x,y)

x = np.array([10,15,20,25,21,24,27,23])
y = np.array([35,32,41,36,34,37,38,30])


plt.scatter(x,y)
plt.show()
```

```python
# WUTH DIFFERENT COLORS

x = np.array([10,15,20,25,21,24,27,23])
y = np.array([35,32,41,36,34,37,38,30])

color = np.array(["black","brown","green","blue","orange","red","pink","purple"])
plt.scatter(x,y,c=color)
plt.show()
```

```python
# CMAP / COLOR GRADINGS

x = np.array([23,12,15,17,13,11,21,18])
y = np.array([44,45,49,55,46,49,54,65])

color = np.array([10,20,30,40,50,55,60,65])
plt.scatter(x,y, c=color, cmap="inferno")
plt.colorbar()
plt.show()
```

```python
# SIZE OF SCATTERS

x = np.array([23,12,15,17,13,11,21,18,20,22,26])
y = np.array([44,45,49,55,46,49,54,65,55,53,46])
size = np.array([50,150,250,350,450,550,650,750,850,950,1050])

plt.scatter(x,y, s=size)
plt.show()
```

```python
# ALPHA SHADING

x = np.array([23,12,15,17,13,11,21,18,20,22,26])
y = np.array([44,45,49,55,46,49,54,65,55,53,46])
size = np.array([50,150,250,350,450,550,650,750,850,950,1050])

plt.scatter(x,y, s=size, c="k", alpha=0.5)
plt.show()
```

```python
# RANDOM SCATTERS

a = np.random.randint(100,size=(100))
b = np.random.randint(100,size=(100))
c = np.random.randint(100,size=(100))
d = 10*np.random.randint(100,size=(100))

plt.scatter(a,b, c=c, s=d, alpha=0.5, cmap="cool")
plt.colorbar()
plt.show()
```
