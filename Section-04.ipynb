```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([3,4,5,6,1,2,5])
y = np.array([1,2,3,4,5,6,7])

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.xticks(y)
plt.plot(y,x)

a = np.array(["Banana","Chickoo","Pineapple"])
b = np.array([10,20,15])
plt.subplot(1,2,2)
plt.bar(a,b)


plt.show()
```

```python
x = np.array([3,4,5,6,1,2,5])
y = np.array([1,2,3,4,5,6,7])

plt.figure(figsize=(5,7))
plt.subplot(2,1,1)
plt.xticks(y)
plt.plot(y,x,color="k")

a = np.array(["Banana","Chickoo","Pineapple"])
b = np.array([10,20,15])
plt.subplot(2,1,2)
plt.bar(a,b,color=["yellow","#696112","lawngreen"])


plt.show()
```

```python
plt.figure(figsize=(20,10))

x = np.array([1,2,3,4])
y = np.array([10,20,30,40])
plt.subplot(2,3,1)
plt.plot(x,y)


x = np.array([10,12,11,8,7,9])
y = np.array([12,13,18,15,11,9])
plt.subplot(2,3,2)
plt.plot(x, linestyle="-.", color="green", linewidth=3)
plt.plot(y, linestyle="dotted", color="maroon", linewidth=3)


a = np.array([9,3,13,18,4,8])
plt.subplot(2,3,3)
plt.plot(a, color="r")


a = np.array([10,6,5,2,7,9,4])
b = np.arange(1,8)
plt.subplot(2,3,4)
plt.plot(b,a,"D-.g",mec="black",ms=10)



s = np.arange(1,6)
e = np.exp(s)
plt.subplot(2,3,5)
plt.plot(s,e,color="black")



a = np.array([10,6,5,2,7,9,4])
b = np.arange(1,8)
plt.subplot(2,3,6)
plt.plot(b,a,"o:r")

plt.show()
```

```python
plt.figure(figsize=(20,10))

x = np.array(["A","B","C","D"])
y = np.array([12,15,10,7])
plt.subplot(2,2,1)
plt.barh(x,y)


fruit = np.array(["Apples","Bananas","Oranges"])
price = np.array([60,40,25])
plt.subplot(2,2,2)
plt.bar(fruit,price,color="orange")


x = np.array(["A","B","C","D"])
y = np.array([8,5,2,7])
plt.subplot(2,2,3)
plt.bar(x,y,color="maroon")


men = [24,21,27,29,40,23]
women = [27,20,18,15,30,17]
groups = np.arange(6)
plt.subplot(2,2,4)
plt.bar(groups,men, color="r", width=.35)
plt.bar(groups,women, color="b", width=.35, bottom = men)
plt.xlabel("Groups")
plt.ylabel("Population")
plt.title("Gender Population In Groups")
plt.xticks(groups,("G1","G2","G3","G4","G5","G6"))
plt.yticks(np.arange(0,81,10))
plt.legend(labels=["Men","Women"])

plt.show()
```
