```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# SIMPLE PIE CHART

x = np.array([30,50,10,10])

plt.pie(x)
plt.show()
```

```python
# PIE CHART WITH LABELS

x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]

plt.pie(x, labels=lb)
plt.show()
```

```python
# CHANGING ANGLE OF PIE CHART

x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]

plt.pie(x, labels=lb, startangle=90)
plt.show()
```

```python
# EXPLODING DIFFERENT PARTS IN A PIE CHART

x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]
e = [0.2,0,0,0]

plt.pie(x, labels=lb, startangle=90, explode=e)
plt.show()
```

```python
x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]
e = [0.1,0.1,0,0]

plt.pie(x, labels=lb, startangle=90, explode=e)
plt.show()
```

```python
x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]
e = [0.1,0.1,0.1,0]

plt.pie(x, labels=lb, startangle=90, explode=e)
plt.show()
```

```python
x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]
e = [0.1,0.1,0.1,0.1]

plt.pie(x, labels=lb, startangle=90, explode=e)
plt.show()
```

```python
# ADDING 3-D EFFECT ON EXPLODED PIE CHART

x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]
e = [0.1,0.1,0,0]

plt.pie(x, labels=lb, startangle=90, explode=e, shadow=True)
plt.show()
```

```python
# ADDING LEGEND TO THE PIE CHART

x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]

plt.pie(x, labels=lb)
plt.legend(title="Vegetables", loc="upper left")
plt.show()
```

```python
x = np.array([30,50,10,10])
lb = ["Tomato","Potato","Ginger","Parsley"]

plt.pie(x)
plt.legend(labels=lb, title="Vegetables", loc="upper left")
plt.show()
```

```python
# ADDING PERCENTAGES IN PIE CHART

lang = np.array(["Python","Java","Swift","C++","C#","JavaScript","R"])
studs = np.array([50,41,13,24,38,47,18])

plt.pie(studs, labels=lang, autopct="%1.3f%%")
plt.show()
```
