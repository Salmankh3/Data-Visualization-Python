```python
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes(projection= "3d")

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

ax.plot3D(x,y,z,"gray")
plt.title("3D Line Plot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes(projection= "3d")

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
c = x + y

ax.scatter(x,y,z,c=c)
plt.title("3D Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes(projection= "3d")

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
c = x - y

ax.scatter(x,y,z,c=c, depthshade=False)
plt.title("3D Scatter Plot")
plt.show()
```

```python
x = np.outer(np.linspace(-2,2,30), np.ones(30))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)

plt.figure(figsize=(8,6))
ax = plt.axes(projection= "3d")

ax.plot_surface(x,y,z,cmap="hot",edgecolor="none")
plt.title("Surface Plot")
plt.show()
```

```python
x = np.outer(np.linspace(1,10,200), np.ones(200))
y = x.copy().T
z = np.cos(x+y)

plt.figure(figsize=(8,6))
ax = plt.axes(projection= "3d", facecolor="gray")

ax.plot_surface(x,y,z,cmap="cool",edgecolor="none")
plt.title("Surface Plot")
plt.show()
```
