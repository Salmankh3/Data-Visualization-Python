# HISTOGRAM

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(170,20,30)
x
```

```python
plt.hist(x)
plt.show()
```

```python
plt.hist(x, histtype="bar")
plt.show()
```

```python
plt.hist(x, histtype="bar", color="k")
plt.show()
```

```python
plt.hist(x, histtype="step", color="r")
plt.show()
```

```python
plt.hist(x, histtype="bar", align="left",color="g")
plt.show()
```

```python
plt.hist(x, histtype="step", orientation="horizontal")
plt.show()
```

```python
plt.hist(x, histtype="bar", align="left",color="orange", density=True)
plt.show()
```
